## Customer Segmentation with RFM
### Business problem
 FLO wants to segment its customers and determine marketing strategies based on these segments. To achieve this, customer behaviors will be identified, and groups will be formed based on these behavior clusters.
 
### Dataset Story
The dataset consists of information derived from the past shopping behaviors of customers who made their last purchases in 2020-2021 through OmniChannel (both online and offline) channels.

master_id: Unique customer number

order_channel: The channel used for shopping (Android, iOS, Desktop, Mobile, Offline)

last_order_channel: The channel used for the last purchase

first_order_date: The date of the customer's first purchase

last_order_date: The date of the customer's last purchase

last_order_date_online: The date of the customer's last online purchase

last_order_date_offline: The date of the customer's last offline purchase

order_num_total_ever_online: Total number of purchases made by the customer online

order_num_total_ever_offline: Total number of purchases made by the customer offline

customer_value_total_ever_offline: Total amount paid by the customer for offline purchases

customer_value_total_ever_online: Total amount paid by the customer for online purchases

interested_in_categories_12: List of categories the customer has shopped in the last 12 months

## Functionize the data preprocessing process.

```python
def data_preparation(df):
    df.head(10)
    df.columns
    df.shape
    df.describe().T
    df.isnull().sum()
    df.info()
    df["total_order_num"] = df["order_num_total_ever_online"] + df["order_num_total_ever_offline"]
    df["total_customer_value"] = df["customer_value_total_ever_offline"] + df["customer_value_total_ever_online"]
    df.head()
    for col in df.columns:
        if "date" in col:
            df[col] = pd.to_datetime(df[col])
    df.info()
    df.sort_values(by="total_customer_value", ascending=False).head(10)
    df.sort_values(by="total_order_num", ascending=False).head(10)
    return df
data_preparation(df)
```

## TASK 2: Calculation of RFM Metrics

```python
df["last_order_date"].max()
today_date = dt.datetime(2021, 6, 1)

rfm = df.groupby("master_id").agg({"last_order_date": lambda date: (today_date - date.max()).days,
                                   "total_order_num": lambda x: x.sum(),
                                   "total_customer_value": lambda x: x.sum()})
rfm.columns = ["recency", "frequency", "monetary"]
```

## TASK 3: Calculation of RF and RFM Scores

```python
rfm["recency_score"] = pd.qcut(rfm["recency"], 5, labels=[5, 4, 3, 2, 1])
rfm["frequency_score"] = pd.qcut(rfm["frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm["monetary_score"] = pd.qcut(rfm["monetary"], 5, labels=[1, 2, 3, 4, 5])
rfm.head()
```

```python
rfm["RF_score"] = (rfm["recency_score"].astype(str) + rfm["frequency_score"].astype(str))
rfm.head()
```

## TASK 4: Definition of RF Scores as Segments

```python
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}

rfm["segment"] = rfm["RF_score"].replace(seg_map,regex=True)
rfm.head()
```

#### 2.Using RFM analysis, identify the relevant customers for 2 cases and save their customer IDs to a CSV file.
a. FLO is introducing a new women's shoe brand. The prices of the products from this brand are above the general customer preferences. Therefore, FLO wants to communicate specifically with the interested customers for the promotion and sales of these products. These customers are planned to be loyal (champions, loyal_customers) and those who shop from the women's category. Save the customer IDs to a CSV file named "new_brand_customer_id.csv".

```python
rfm.reset_index(inplace=True)
target_segment = rfm[rfm["segment"].isin(["champions", "loyal_customers"])]["master_id"]
customer_ids = df[(df["master_id"].isin(target_segment)) & (df["interested_in_categories_12"].str.contains("KADIN"))]["master_id"]
customer_ids.to_csv("new_brand_customer_id.csv", index=False)
customer_ids.shape
```
